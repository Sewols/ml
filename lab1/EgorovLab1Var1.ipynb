**6. Какое самое популярное женское имя на корабле? Извлеките из полного имени пассажира (колонка Name) его личное имя (First Name). Попробуйте вручную разобрать несколько значений столбца Name и выработать правило для извлечения имен, а также разделения их на женские и мужские.**

 - 'Mrs.' - женщина замужем(Имя в круглых скобках)  
 - 'Miss.' - женщина не замужем.

```python
female = (titanic[titanic["Sex"] == "female"])["Name"].reset_index(drop=True)
female = female.apply(lambda x: x.split(',')[1].split('.')[1])
for i, name in enumerate(female):
    if name.find("(") != -1:
        female[i] = name.split("(")[1]
female = female.apply(lambda x: x.strip(' ').split(' ')[0])
names_rate_list = female.value_counts()
print("The name ", names_rate_list.idxmax(), "meets ", names_rate_list.max() , "times")
```

**8. Визуализируйте гистограммы возраста для выживших и не выживших пассажиров. Сделайте выводы. Отобразите данные на одном и нескольких графиках**

```python
alive = titanic[titanic['Survived'] == True]
died = titanic[titanic['Survived'] == False]

alive_ages = alive['Age']
died_ages = died['Age']

alive_ages = alive_ages.dropna().reset_index(drop=True)
died_ages = died_ages.dropna().reset_index(drop=True)

sns.distplot(died_ages, bins=10, hist=True, label='Not surviving', kde=False)
sns.distplot(alive_ages, bins=10, hist=True, label='Survived', kde=False)

plt.legend()
plt.show()
```

**9. Визуализируйте гистограммы возраста для выживших и не выживших пассажиров по классам.**

```python

for cls in range(1, 4):
    title = 'Class №' + str(cls)
    
    for surv in range(2):
        if surv == True:
            event = 'Survived'
        else:
            event = "Not survived"
            
        temp = titanic[titanic['Survived'] == surv]
        temp = temp[temp['Pclass'] == cls]
        temp = temp['Age']
        
        temp = temp.dropna().reset_index(drop=True)
        sns.distplot(temp, bins=10, hist=True, label=event, kde=False)
        
    plt.title(title)
    plt.legend()
    plt.show()
```

**10. Постройте столбчатую диаграмму количества людей: мужчины, женщины, дети.**
